```python
from keras.models import load_model
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
X,y = load_breast_cancer(return_X_y=True, as_frame=True)
import tensorflow as tf
scaler = StandardScaler()
#split the data 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 4)
X_train_scale = pd.DataFrame(scaler.fit_transform(X_train),columns = X.columns)
X_test_scale = pd.DataFrame(scaler.transform(X_test),columns = X.columns)
```

```python
model = load_model('ann_clf.h5')
model.summary()
```

```python
model.add(tf.keras.layers.Lambda(lambda l: l[:, 1]))
model.summary()
model.save('edited_model.h5')
```

```python
Analysis_ = Analysis(X_train_scale,X_test_scale,y_train,y_test, 'edited_model.h5')
Analysis_.feature_importance_values()
```
